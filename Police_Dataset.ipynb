#### Q. 4) Question ( mapping + data-type casting ) 
- What is the mean stop_duration ?

```python
police_data["stop_duration"].value_counts()
```

```python
police_data["stop_duration"] = police_data["stop_duration"].map({"0-15 Min" : 7.5, "16-30 Min": 23, "30+ Min" : 45 })
```

```python
police_data['stop_duration'].value_counts().plot(kind="bar", color='green')
plt.title("Count of Stop_Duration")
```

```python
police_data
```

```python
police_data["stop_duration"].mean()
```
